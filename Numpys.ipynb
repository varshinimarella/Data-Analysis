```python
import numpy as np
import pandas as pd

```

```python
products = pd.read_csv("products.csv")
customers = pd.read_csv("customers.csv")
employees = pd.read_csv("employees.csv")
```

```python
#Convert colums to numpy arrays
price = products["Price"].values
category_id = products["CategoryID"].values
customer_city = customers["CityID"].values
```

```python
#Numpy array properties
print("Price Shape:", price.shape)
print("Price Size:", price.size)
print("Price Data Type:", price.dtype)
```

```python
#statistical opearations using numpys
print("Maximum Price:", np.max(price))
print("Minimum Price:", np.min(price))
print("Mean Price:", np.mean(price))
print("Total Price Sum:", np.sum(price))
```

```python
#conditional filtering using numpys
above_avg_price = price[price > np.mean(price)]
above_avg_price[:10]
```

```python
#Discount caluclation using numpys
discount = np.random.choice([0, 0.05, 0.1, 0.2], size=len(price))
discounted_price = price * (1 - discount)
discounted_price[:10]
```

```python
#Quamtity and revenue caluclations using numpys 
quantity = np.random.randint(1, 10, size=len(price))
revenue = discounted_price * quantity

print("Total Revenue:", np.sum(revenue))
print("Average Revenue:", np.mean(revenue))
```

```python
#Sorting Revenue
sorted_revenue = np.sort(revenue)
sorted_revenue[-10:]
```

```python
#category wise Revenue
unique_categories = np.unique(category_id)
category_revenue = {}

for cat in unique_categories:
    category_revenue[cat] = np.sum(revenue[category_id == cat])

category_revenue
```

```python
#Reshaping numpy arrays
revenue_matrix = revenue[:20].reshape(4, 5)
revenue_matrix
```
