```python
#Upload / Load CSV Files (Pandas Code)
categories = pd.read_csv("categories.csv")
products = pd.read_csv("products.csv")
categories = pd.read_csv("categories.csv")
customers = pd.read_csv("customers.csv")
employees = pd.read_csv("employees.csv")
cities = pd.read_csv("cities.csv")
countries = pd.read_csv("countries.csv")
```

```python
def data_overview(df, name):
    print(f"\n--- {name} Dataset ---")
    print("Shape:", df.shape)
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum())
```

```python
data_overview(products, "Products")
data_overview(categories, "Categories")
data_overview(customers, "Customers")
data_overview(employees, "Employees")
data_overview(cities, "Cities")
data_overview(countries, "Countries")
```

```python
#Remove Duplicates
products.drop_duplicates(inplace=True)
categories.drop_duplicates(inplace=True)
customers.drop_duplicates(inplace=True)
employees.drop_duplicates(inplace=True)
cities.drop_duplicates(inplace=True)
countries.drop_duplicates(inplace=True)
```

```python
#handle Missing Values
products = products.fillna({
    "VitalityDays": 0,
    "IsAllergic": "Unknown",
    "Resistant": "Unknown"
})

```

```python
customers.fillna({
    "Gender": "Unknown",
}, inplace=True)
```

```python
cities.fillna("Unknown", inplace=True)
countries.fillna("Unknown", inplace=True)
```

```python
#Convert Date Columns to DateTime
products["ModifyDate"] = pd.to_datetime(products["ModifyDate"], errors="coerce")
```

```python

customers.columns

```

```python
# Remove leading/trailing spaces and make consistent
customers.columns = customers.columns.str.strip()
customers.columns = customers.columns.str.replace(" ", "_")  # optional
print(customers.columns)
```

```python
# Check for similar names
[col for col in customers.columns if "date" in col.lower()]



```

```python
employees["HireDate"] = pd.to_datetime(employees["HireDate"], errors="coerce")
```

```python
#rename Columns
products.columns = products.columns.str.lower()
categories.columns = categories.columns.str.lower()
customers.columns = customers.columns.str.lower()
employees.columns = employees.columns.str.lower()
cities.columns = cities.columns.str.lower()
countries.columns = countries.columns.str.lower()
```

```python
#Validate key columns
# Customers – CityID validation
assert customers["cityid"].notnull().all()
assert customers["cityid"].dtype in [np.int64, np.int32]

# Employees – CityID validation
assert employees["cityid"].notnull().all()
assert employees["cityid"].dtype in [np.int64, np.int32]

```

```python
#category validation
products = products[products["categoryid"].isin(categories["categoryid"])]
```

```python
#city validation
customers = customers[customers["cityid"].isin(cities["cityid"])]
employees = employees[employees["cityid"].isin(cities["cityid"])]
```

```python
#final cleaned data set
data_overview(products, "Cleaned Products")
data_overview(customers, "Cleaned Customers")
data_overview(employees, "Cleaned Employees")
```

```python
products.to_csv("clean_products.csv", index=False)
categories.to_csv("clean_categories.csv", index=False)
customers.to_csv("clean_customers.csv", index=False)
employees.to_csv("clean_employees.csv", index=False)
cities.to_csv("clean_cities.csv", index=False)
countries.to_csv("clean_countries.csv", index=False)
```
