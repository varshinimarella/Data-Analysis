```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
products = pd.read_csv('products.csv')

products['ModifyDate'] = pd.to_datetime(products['ModifyDate'], errors='coerce')
products = products.dropna(subset=['ModifyDate'])

products['Price'] = products['Price'].fillna(products['Price'].mean())
```

```python
monthly_price = (
    products.groupby(products['ModifyDate'].dt.to_period('M'))['Price']
            .mean()
            .reset_index()
)

monthly_price['ModifyDate'] = monthly_price['ModifyDate'].dt.to_timestamp()
```

```python

import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(layout="wide")
st.title("ðŸ“Š Product Analytics Dashboard")

products = pd.read_csv('products.csv')

products['ModifyDate'] = pd.to_datetime(products['ModifyDate'], errors='coerce')
products = products.dropna(subset=['ModifyDate'])
products['Price'] = products['Price'].fillna(products['Price'].mean())

monthly_price = (
    products.groupby(products['ModifyDate'].dt.to_period('M'))['Price']
            .mean()
            .reset_index()
)

monthly_price['ModifyDate'] = monthly_price['ModifyDate'].dt.to_timestamp()

st.subheader("ðŸ“ˆ Monthly Average Product Price Trend")

fig = px.line(
    monthly_price,
    x='ModifyDate',
    y='Price',
    markers=True
)

st.plotly_chart(fig, use_container_width=True)
```
